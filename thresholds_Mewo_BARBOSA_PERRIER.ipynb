## Pre-process data

```python
# 1 : reduce noise (value < 0.01 -> 0)
# low probabilities may noise data
x_train[x_train < 0.01] = 0
x_train.sample(n=10)
```

```python
lowest_count_for_a_tag = 999999
low_tag = None
highest_count_for_a_tag = 0
high_tag = None

for col in y_train:
    if col != 'ChallengeID':
        col_sum = y_train[col].sum()
        #print('Count of ' + col + " tag : "+ str(col_sum))
        if col_sum < lowest_count_for_a_tag:
            lowest_count_for_a_tag = col_sum
            low_tag = col
        if col_sum > highest_count_for_a_tag:
            highest_count_for_a_tag = col_sum
            high_tag = col
            
print("Lowest count for a tag = " + str(lowest_count_for_a_tag) + ' for tag : ' + low_tag)
print("Highest count for a tag = " + str(highest_count_for_a_tag) + ' for tag : ' + high_tag)
```

# Approach 2 : treshold for each tag (tresholding column wise)

```python
# !!! VERY long to compute...

# calculate f1 scores
def get_scores(thresholds, input_col, truth_col):
    scores = []
    for thresh in thresholds:
        true_pos = 0
        false_pos = 0
        false_neg = 0
        for i in range(len(input_col)):
            if input_col[i] >= thresh and truth_col[i] == 1:
                true_pos += 1
            if input_col[i] >= thresh and truth_col[i] == 0:
                false_pos -= 1
            if input_col[i] < thresh and truth_col[i] == 1:
                false_neg += 1
        score = true_pos/(true_pos+0.5*(false_pos+false_neg))
        scores.append(score)
    return scores

def benchmark_thresholding(input_data, labels, truth_values):
    output_data = input_data[:][labels.columns[:]]
    thresholds = np.arange(0, 0.6, 0.01)
    output_thresholds = []
    for col in output_data.columns:
        output_data[col].values[:] = 0 #init output_data
    for j in labels.columns[:]:
        scores = get_scores(thresholds, input_data[j], truth_values[j])
        thresh = thresholds[np.argmax(scores)]
        print(j, thresh)
        output_thresholds.append(thresh)
        #output_data[j][:] = get_indices(thresh, input_data[j]).values
    return output_thresholds

def fixed_thresholding(input_data, labels, thresholds):
    output_data = input_data[:][labels.columns[:]].copy()
    for col in output_data.columns:
        output_data[col].values[:] = 0 #init output_data
    for i,j in enumerate(labels.columns[:]):
        index = input_data.index[input_data[j] > thresholds[i]]
        output_data[j].loc[index] = 1
    return output_data

labels = pd.read_csv('annex/mewo-labels.csv', index_col=0, sep=',')
thresholds = benchmark_thresholding(x_train, labels, y_train)
df_y_test = fixed_thresholding(x_test, labels, thresholds)
df_y_test.to_csv('pred_y_th.csv')
# score : 0.3945
```
